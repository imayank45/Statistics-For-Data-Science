```python
from scipy import stats
import seaborn as sns
```

```python
class_A = [78, 85, 90, 76, 89, 84, 92, 88, 79, 81, 95, 87, 80, 86, 91]
class_B = [82, 79, 88, 84, 90, 85, 89, 93, 77, 80, 94, 86, 83, 87, 92]
```

```python
sns.distplot(class_A, label='Class A')
sns.distplot(class_B, label='Class B')
```

```python
# calculate variances
print("Var1", stats.variation(class_A))
print("Var2", stats.variation(class_B))
```

```python
# Welch's t-test (assumes unqual variance)
t_test, p_value = stats.ttest_ind(class_A, class_B, equal_var=False)
```

```python
# Output results
print(f"t-statistic: {t_test:.4f}, p-value: {p_value:.4f}")
```

```python
# Hypothesis testing interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in means.")
else:
    print("Fail to reject the null hypothesis: No significant difference in means.")
```

```python
# Pooled t-test
t_test, p_value = stats.ttest_ind(class_A, class_B, equal_var=True)
```

```python
# Output results
print(f"t-statistic: {t_test:.4f}, p-value: {p_value:.4f}")
```

```python
# Hypothesis testing interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in means.")
else:
    print("Fail to reject the null hypothesis: No significant difference in means.")
```
